```python

# Get the data out of the dataframe into a numpy matrix and keep only the name and gender columns
#dataset = dataset.as_matrix()[:,0]
#print (dataset)
 
# We're using 80% of the data for training
TRAIN_SPLIT = 0.8
```

```python
def features(language):
  #  name = name.lower()
    return {
        'first-letter': language[0], # First letter
        'first2-letters': language[0:2], # First 2 letters
        'first3-letters': language[0:3], # First 3 letters
        'last-letter': language[-1],
        'last2-letters': language[-2:],
        'last3-letters': language[-3:],
        'last4-letters': language[-4:],
    }
 
print (features("accessorise"))
# {'first2-letters': 'jo', 'last-letter': 'n', 'first-letter': 'j', 'last2-letters': 'hn', 'last3-letters': 'ohn', 'first3-letters': 'joh'}
 
```

```python
# Vectorize the features function
features = np.vectorize(features)
#print (features(["aesthetic", "acclimatising", "agonise"]))

# Extract the features for the whole dataset
X = features(dataset[0]) # X contains the features
#print(X[100]) 
# Get the country column
y = dataset[1]           # y contains the targets
 
# Test if we built the dataset correctly
print ("Name: %s, features=%s, gender=%s" % (dataset[0][0], X[0], y[0]))
# Name: Mary, features={'first2-letters': 'ma', 'last-letter': 'y', 'first-letter': 'm', 'last
```

```python
from sklearn.utils import shuffle
X, y = shuffle(X, y)
X_train, X_test = X[:int(TRAIN_SPLIT * len(X))], X[int(TRAIN_SPLIT * len(X)):]
y_train, y_test = y[:int(TRAIN_SPLIT * len(y))], y[int(TRAIN_SPLIT * len(y)):]
 
# Check to see if the datasets add up
print (len(X_train), len(X_test), len(y_train), len(y_test))  
```

```python
from sklearn.feature_extraction import DictVectorizer
 
print (features(["accessorise", "accessorize"]))
vectorizer = DictVectorizer()
vectorizer.fit(X_train)
 
transformed = vectorizer.transform(features(["accessorise", "accessorize"]))
print (transformed)
```

```python
from sklearn.tree import DecisionTreeClassifier
 
clf = DecisionTreeClassifier()
clf.fit(vectorizer.transform(X_train), y_train)
```

```python
# Accuracy on training set
print (clf.score(vectorizer.transform(X_train), y_train)) 
 
# Accuracy on test set
print (clf.score(vectorizer.transform(X_test), y_test))   
 
```
