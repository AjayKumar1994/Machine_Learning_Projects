```python
# replacing NAN values with 0
train=train.fillna(value=0)
train
```

```python
# we use spearman correlation as it can use categorical as well as continuous data
# Assumption : monotonic relationship (either variables inc. together or one inc. and other decreases)
train.corr(method='spearman')
```

```python
# Taking different features as independent variables(x) to predict the shape(y)
#Only considered the mandatory features for the analysis
x=train[['Length','Width','Height','Diameter','Method of Manufacture_Extrusion BM','Method of Manufacture_Inject-Stretch BM','Method of Manufacture_Injection BM','Method of Manufacture_Injection Moulded','Neck Finish-Bottle_24mm','Neck Finish-Bottle_28mm','Neck Finish-Bottle_33mm','Neck Finish-Bottle_40mm','Neck Finish-Bottle_Other','Neck Finish Style No.1_External Threaded','Neck Finish Style No.1_Multi-thread','Neck Finish Style No.1_Single thread','Neck Finish Style No.1_Snap-on']]

y= train[['Type_Cylindrical','Type_Other','Type_Oval','Type_Rectangular','Type_Round','Type_Round cnr rectangle']]
```

```python
# import modules for Decision tree
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```
